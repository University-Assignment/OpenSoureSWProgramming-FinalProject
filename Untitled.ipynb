```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
from numpy import NaN


plt.rcParams['font.family'] = 'Gulim'


class Movie:
    def __init__(self, title, star, rating, year, time, count, genre, ages):
        self.title = title
        self.star = star
        self.rating = rating
        self.year = year
        self.time = time
        self.count = count
        self.ages = ages
        self.genre = genre
        
    def show(self):
        print('Title: ' + str(self.title))
        print('Star: ' + str(self.star))
        print('Rating: ' + str(self.rating))
        print('Year: ' + str(self.year))
        print('Time: ' + str(self.time))
        print('Count: ' + str(self.count))
        print('Genre: ' + str(self.genre))
        for i in range(len(self.ages)):
            print(str((i + 1) * 10) + '대: ' + str(self.ages[i]))
  
```

```python
def crawlSubData(url):
    
    soup = BeautifulSoup(urllib.request.urlopen(url).read(), "html.parser")
    review_list = []
    age_list = []
    title = soup.find('h3', class_='h_movie').find('a').text
    graph = soup.find("div", class_="bar_graph")
    if graph != None:
        age = graph.select("div > strong")
        for index in range(5):
            age_list.append(str(age[index * 2].text).strip('%'))
    else:
        for index in range(5):
            age_list.append(NaN)
    count = soup.find("p", class_="count")
    if count == None:
        count = NaN
    else:
        count = str(count.text).split("명")[0].replace(',', '')
    
    return count, age_list
```

```python
def getSummary(movie_list):
    title_list = []
    star_list = []
    rating_list = []
    year_list = []
    time_list = []
    count_list = []
    genre_list = []
    age_list = []
    
    for movie in movie_list:
        title_list.append(movie.title)
        star_list.append(movie.star)
        rating_list.append(movie.rating)
        year_list.append(movie.year)
        time_list.append(movie.time)
        count_list.append(movie.count)
        genre_list.append(movie.genre)
        if movie.ages[0] != NaN:
            max = 0
            for i in range(4):
                if movie.ages[i + 1] > movie.ages[max]:
                    max = i + 1
            age_list.append((max + 1) * 10)
        else:
            age_list.append(NaN)

    
    title_series = pd.Series(title_list)
    star_series = pd.Series(star_list)
    rating_series = pd.Series(rating_list)
    year_series = pd.Series(year_list)
    time_series = pd.Series(time_list)
    count_series = pd.Series(count_list)
    genre_series = pd.Series(genre_list)
    age_series = pd.Series(age_list)
  
    summary = pd.DataFrame({
      'Title': title_series,
      'Star': star_series,
      'Rating': rating_series,
      'Year': year_series,
      'Time': time_series,
      'Count': count_series,
      'Genre': genre_series,
      'Age': age_series
    })
  
    return summary
```

```python
soup = BeautifulSoup(urllib.request.urlopen("https://movie.naver.com/movie/running/current.nhn").read(), "html.parser")
movies = soup.find("ul", class_="lst_detail_t1")
head = movies.select("dt > a")
stars = movies.find_all("dd", class_="star")
info_txt1 = movies.find_all("dl", class_="info_txt1")
movie_list = []

for i in range(len(head)):
    url = str(head[i]).split("\">")[0].split("href=\"")[1]
    title = head[i].text
    info_exp = stars[i].find("dl", class_="info_exp")
    star = stars[i].find("dl", class_="info_star").find("span", class_="num").text
    if info_exp != None:
        rating = str(info_exp.find("span", class_="num").text)
    else:
        rating = NaN
    info = str(info_txt1[i]).split("분")
    time = info[0].split("|</span>")
    if len(time) > 1:
        time = time[1].strip()
    else:
        time = info[0].split("</span>")[1].strip()
    year = info[1].split("</span>")[1].split("개봉")[0].strip().split(".")[0]
    count, age_list = crawlSubData("https://movie.naver.com" + str(url))
    genre = info_txt1[i].find("span", class_="link_txt").text.strip()
    if genre.count(',') > 0:
        genre = genre.split(',')[0]
    movie = Movie(title, star, rating, year, time, count, genre, age_list)
    #movie.show()
    movie_list.append(movie)
    
summary = getSummary(movie_list)
summary.to_csv("MovieDataSet.csv", index=False)
```

```python
def getFreqDataFrame(df, name):
    test = df.groupby('Age')[name]
    max = []
    age_list = []
    max_list = []
    unique = test.unique()
    for i in range(10, 51, 10):
        age_list.append(i)
        max.append(0)
        check = []
        for j in range(len(unique[i])):
            check.append(0)
            value = unique[i][j]
            if value != NaN:
                for a, b in test:
                    if a == i:
                        for c in b:
                            if c == str(value):
                                check[j] = check[j] + 1
            if max[int(i / 10 - 1)] < check[j]:
                max[int(i / 10 - 1)] = j
        max_list.append(unique[i][max[int(i / 10 - 1)]])


    age_series = pd.Series(age_list)
    max_series = pd.Series(max_list)

    summary = pd.DataFrame({
        'Age': age_series,
        name: max_series
    })
  
    return summary
```
